```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(10,5))
ct=dataset1['salary'].value_counts()
gen=dataset1['Department'].value_counts()
ax1.bar(x=ct.index, height=ct.values)
ax1.set_title('number_project', fontsize=12)
ax2.bar(x=gen.index, height=gen.values)
ax2.set_title('average_montly_hours', fontsize=12)
```
