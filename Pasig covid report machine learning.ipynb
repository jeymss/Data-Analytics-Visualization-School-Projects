```python
#import all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#1. Convert your PT1A activity using Python
df = pd.read_csv('/content/drive/MyDrive/AI/pasigcovid.csv')
df
```

```python
df.dtypes
```

```python
#Question1. Apply the statistical function of excel:   regression and correlation
#correlation pearson
df.corr(method ='pearson')
```

```python
#correlation kendall
df.corr(method='kendall')
```

```python
#regression
X = df[["CONFIRMD CASES"]]
y = df[["RECOVRED CASES"]]

regressor = LinearRegression()
regressor.fit(X, y)

plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Confirmed Cases vs Recovered Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Recovered Cases')
plt.show()
```

```python
X = df[["ACTIVE CASES"]]
y = df[["RECOVRED CASES"]]

regressor = LinearRegression()
regressor.fit(X, y)

plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Active Cases vs Recovered Cases')
plt.xlabel('Active Cases')
plt.ylabel('Recovered Cases')
plt.show()
```
